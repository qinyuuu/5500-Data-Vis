```python
# setting
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data
pokemon_data = pd.read_csv("Pokemon.csv")
```

```python
# first visualization
plt.hist(pokemon_data["HP"], bins=20)
plt.xlabel('HP value')
plt.ylabel('Number of pokemons')
plt.title("HP value distribution among pokemons")
plt.show()
```

```python
# second visualization
plt.boxplot(pokemon_data["HP"])
plt.ylabel("HP value")
plt.title("HP value distribution among pokemons")
plt.show()
```


(1) 1-2 encodings you decided not to use, and why:
The task is to help users understand the HP value distribution among 700+ pokemon. There is only one continuous variable need to visualize, of which the data type is quantitative. According to the "Channel Ranking by Data Type" on the data type slide, I decided not to use texture, connection, containment, and shape as the channel to encode this quantitative data.

(2) Advantages and disadvantages of each of the encodings you chose:
For the first visualization, the advantage is that we can easily see that the HP value is normally distributed. The disadvantage is that we can not clearly determine the statistical values of the distribution.
For the second visualization, the advantage is that we can clearly know the min, max, median of the HP Value. The disadvantage is that we cannot determine the distribution type from the encodings. 

(3) Which one of the two visualizations you prefer, and why:
I prefer the first encoding. Since the task is to help users understand the HP value distribution. The visualization design needs to address the task. So I think the first visualization performs better.
